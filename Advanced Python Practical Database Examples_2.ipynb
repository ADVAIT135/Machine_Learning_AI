### D. Connect to a DB using SQLALCHEMY within Jupyter Notebook 

```python
from sqlalchemy import create_engine
from os import environ
```

```python
db_url = environ.get('SALESPEOPLE_DB_URL')
engine = create_engine("sqlite:///salespeople.sqlite")
```

### E. Visualize data using SQL Alchemy, pandas, and Matplotlib 

```python
sales_df = pd.read_sql_table("salespeople", con = engine.connect())
#index_col = 'idColumn', coerce_float = True, columns = ["city", "state"]
#parse_dates = ["dateColumn"], chunksize = 250
```
